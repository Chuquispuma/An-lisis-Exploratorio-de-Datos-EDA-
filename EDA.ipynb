### Visualización de datos

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración de estilo de gráficos de seaborn
sns.set(style="whitegrid")
```

```python
# Histograma para una variable numérica como ejemplo (AGE)
sns.histplot(df['AGE'], kde=True)
plt.title('Distribución de la edad')
plt.show()
```

```python
# Gráfico de barras para la variable GENDER
sns.countplot(x='GENDER', data=df)
plt.title('Distribución de género')
plt.show()

# Gráfico de barras para la variable LUNG_CANCER
sns.countplot(x='LUNG_CANCER', data=df)
plt.title('Frecuencia de cáncer de pulmón')
plt.show()
```

```python
# Boxplot para analizar la relación entre la edad y la presencia de cáncer de pulmón
sns.boxplot(x='LUNG_CANCER', y='AGE', data=df)
plt.title('Relación entre edad y cáncer de pulmón')
plt.show()
```

```python
# Boxplot para detectar outliers en la edad
sns.boxplot(x=df['AGE'])
plt.title('Detección de outliers en la edad')
plt.show()
```

```python
# Seleccionar solo columnas numéricas (excluyendo 'GENDER' y 'LUNG_CANCER')
numeric_columns = df.select_dtypes(include=['number']).columns
df_numeric = df[numeric_columns]

# Calcular la correlación solo para las columnas numéricas
correlation_matrix = df_numeric.corr()

# Crear el mapa de calor
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriz de correlación (variables numéricas)')
plt.show()
```
